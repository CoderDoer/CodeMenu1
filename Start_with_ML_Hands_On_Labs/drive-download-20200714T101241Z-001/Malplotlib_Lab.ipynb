# Pyplot :-
collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc

```python
# Providing single list to the array
# Pyplot assumes it as y-axis and x data becomes a range from 0 to 4 [0,1,2,3].
import matplotlib.pyplot as plt
#plt.plot([1,2,3,4])
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'k--')
# Third optional argument ro stands for red dots. b- stands for solid blue line
#plt.axis([0, 20, 0, 20])
# The axis() command in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0, 5,0.20)

print(list(t))

# red dashes, blue squares and green triangles
#plt.plot(t, t, 'r', t, t**2, 'bs', t, t**3, 'g^',t,t**4,'y.',t,t**5,'c--',linewidth=10)
plt.plot(t, t, 'r',linewidth=10)
plt.plot( t, t**2, 'bs')
plt.plot(t, t**3, 'g^',t,t**4,'y-',t,t**5,'c--', linewidth =10)
plt.axis([0,5,0,200])
plt.show()   
```

```python
# Working with multiple figures and axes
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)


plt.subplot(321)
plt.plot(t1, f(t1), 'yo', t2, f(t2), 'k')
plt.title("Subplot 1")
plt.xlabel("x label")
plt.ylabel("y label")

plt.subplot(322)
plt.title("Subplot 2")
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.xlabel("x label")
plt.ylabel("y label")



plt.subplot(323)
plt.title("Subplot 3")
plt.plot(t2, np.cos(2*np.pi*t2), 'g--')
plt.xlabel("x label")
plt.ylabel("y label")


plt.subplot(324)
plt.title("Subplot 4")
plt.plot(t2, np.cos(2*np.pi*t2), 'y--')
plt.xlabel("x label")
plt.ylabel("y label")

plt.subplot(325)
plt.title("Subplot 5")
plt.plot(t2, np.cos(2*np.pi*t2), 'b--')
plt.xlabel("x label")
plt.ylabel("y label")


plt.subplot(326)
plt.title("Subplot 6")
plt.plot(t2, np.cos(2*np.pi*t2), 'c--')
plt.xlabel("x label")
plt.ylabel("y label")

plt.subplots_adjust(wspace=1.5,hspace=1.5)
plt.show()


```

The figure() command here is optional because figure(1) will be created by default, just as a subplot(111) will be created by default if you don’t manually specify any axes. The subplot() command specifies numrows, numcols, fignum where fignum ranges from 1 to numrows*numcols. The commas in the subplot command are optional if numrows*numcols<10. So subplot(211) is identical to subplot(2, 1, 1). You can create an arbitrary number of subplots and axes. If you want to place an axes manually, i.e., not on a rectangular grid, use the axes() command, which allows you to specify the location as axes([left, bottom, width, height]) where all values are in fractional (0 to 1) coordinates. 

```python
# You can create multiple figures by using multiple figure() calls with 
# an increasing figure number.
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
x = [0,1,2]
y=[1,2,3]
plt.plot(y)


plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])
plt.subplots_adjust(wspace=1.5,hspace=1.5)

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3')

plt.subplot(212) 
plt.title('Easy as 4, 5, 6')# subplot 211 title
plt.show()
```
