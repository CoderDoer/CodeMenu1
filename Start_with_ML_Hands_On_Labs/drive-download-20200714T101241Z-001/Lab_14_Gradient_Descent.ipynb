### GRADIENT DESCENT 

_Description : It is an optimization algorithm used to find the minimum of a function. We start with a random point on the function and move in the negative direction of the gradient of the function to reach the local/global minima._

__Approach : You can think of the optimization algorithm of the cost function as a blind folded man standing at some point of a mount range trying to find the lake at the bottom. Try to Think of the indicators that can quickly lead him to the local/global minima.__  

```python
from IPython.display import Image
Image(filename='imgr.png',width=300, height=300) 
```

  __Note__:   `y = (x+5)² ` reaches it’s minimum value when x = -5 (i.e when x=-5, y=0). 
               Hence x=-5 is the local and global minima of the function.

-  Initialize x =3 randomly.
-  Then, find the gradient of the function, dy/dx = 2*(x+5).
-  Move in the direction of the negative of the gradient (we are looking for the
   update to get us smaller value than current x)
-  Learning rate → 0.01 decides how much to move. 
-  Let’s perform 2 iterations of gradient descent

```python
from IPython.display import Image
Image(filename='imgr2.png',width=500, height=500) 
```

`We can observe that the X value is slowly decreasing and should converge to -5 (the local minima).`

However, how many iterations should we perform?

This can be decided by setting a __Precision Variable__

If the difference between x values from 2 consecutive iterations is lesser than the precision we set, stop the algorithm

__Step 1 : Initialize parameters__

```python
cur_x = 3 
# The algorithm starts at x=3
```

```python
rate = 0.01 
# Learning rate
```

```python
precision = 0.000001 
#This tells us when to stop the algorithm
```

```python
previous_step_size = 1 
```

```python
max_iters = 10000 
# maximum number of iterations
```

```python
iters = 0 
#iteration counter
```

```python
df = lambda x: 2*(x+5) 
#Gradient of our function
```

__Step 2 : Run a loop to perform gradient descent__

`Stop loop when difference between x values from 2 consecutive iterations is less than 0.000001 or when number of iterations exceeds 10,000`

```python
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    print("Iteration",iters,"\n Updated X value is",cur_x) #Print iterations
    print("\n")
    
print("The local minimum occurs at", cur_x)
```

_The algorithm runs for 595 iterations before it terminates._
